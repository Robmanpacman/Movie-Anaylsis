```python
#importing libraries that are neccessary
from bs4 import BeautifulSoup as bs
import requests,openpyxl
```

```python
#grabbing url
r = requests.get('https://www.imdb.com/search/title/?groups=top_250&sort=boxoffice_gross_us,desc')

#making url into beautiful soup object
soup = bs(r.content)

#using prettify to make html code readable
contents = soup.prettify()

#confirming html code
print(contents)
```

```python
#finding the specific info i want from the html code using BeautifulSoup
try:
    movies = soup.find_all('h3')
    for movie in movies:
        print(movie.prettify())
except Exception as e:
    print(e)
```

```python
#extracting the specific info i want into a dictionary list
movie_info = {}
for index,row in enumerate (movies):
    if index == 0:
        movie_info['title'] = row.find("h3").text(" ",strip=True)

movie_info
```
